```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#cleaning up data
```

```python
#only want 50km or 50 mi
```

```python
df[df['Event distance/length'] == '50km']
```

```python
df[df['Event distance/length'] == '50mi']
```

```python
#now we have to combine both 50 miles and 50 Km
```

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
#this way we have all the races in USA 
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA' ]
```

```python
#combine all data filters
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2
```

```python
df2.shape
```

```python
#removing usa from eventsname
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# lets remove the 'h' from that athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(10)
```

```python
#lets drop columns that arent important
```

```python
# athlete club, country, year of birth, age category
```

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

```python
df2.head(10)
```

```python
# lets check for null values in our table
```

```python
df2.isna().sum()
```

```python
# no null values--- our data is clean
```

```python
# lets check for duplicate values
```

```python
df2[df2.duplicated() == True]
```

```python
# we have to drop duplicated data
```

```python
df2 = df2.drop_duplicates()
```

```python
df2.shape
```

```python
# reset index
```

```python
df2.reset_index(drop = True)
```

```python
#since the data is clean... we can state the visualization
```

```python
sns.boxplot(data=df2, x='Athlete gender', y='Year of event')

plt.show()
```

```python

# Create the scatter plot
sns.scatterplot(data=df2, x='Athlete gender', y='Year of event')
plt.show()
```

```python
data = df2['Year of event'].value_counts()

# Filter data to start from 1985 to 1999
filtered_data = data[(data.index >= 1985) & (data.index <= 1999)]

# Create a pie chart
plt.figure(figsize=(20, 20))
plt.pie(filtered_data, labels=filtered_data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Column Values Starting from 1985')
plt.show()
```

```python
# Assuming 'Year of event' is the column you want to visualize
# You may replace it with the actual column name in your DataFrame
data = df2['Year of event'].value_counts()

# Filter data to start from 2000 to 2024
filtered_data = data[(data.index >= 2007) & (data.index <= 2024)]

# Create a pie chart
plt.figure(figsize=(20, 20))
plt.pie(filtered_data, labels=filtered_data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Column Values Starting from 2007')
plt.show()

```
